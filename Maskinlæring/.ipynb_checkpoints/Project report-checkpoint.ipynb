```python
import pandas as pd
import numpy as np
import os

import h2o
from h2o.estimators.glm import H2OGeneralizedLinearEstimator
from h2o.estimators.gbm import H2OGradientBoostingEstimator
from h2o.estimators.random_forest import H2ORandomForestEstimator
from h2o.estimators.deeplearning import H2ODeepLearningEstimator
from h2o.grid.grid_search import H2OGridSearch

from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import cross_val_score, RandomizedSearchCV
from sklearn import metrics

import warnings
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def ROC_Curve(model, df):
    performance = model.model_performance(df)
    auc = performance.auc()
    false_positive_rate = performance.fprs
    true_positive_rate = performance.tprs

    plt.style.use('ggplot')
    plt.figure()
    plt.plot(false_positive_rate, true_positive_rate, 'k--')
    plt.plot(false_positive_rate, 
             true_positive_rate, 
             color='darkorange',
             lw = 2,
             label='ROC curve (area = %0.2f)' % auc)
    plt.plot([0,1], [0,1], color = 'navy', lw = 2, linestyle = '--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False positive rate')
    plt.ylabel('True positive rate')
    plt.title('ROC curve')
    plt.legend(loc='best')
    plt.show()
    


```

## Data preparation

While H2O handles most of te data processing natively the models built using scikit-learn require some steps of data preparation.

```python
x = train_h2o.columns
x.remove("target")
x.remove("id")
y = "target"
```

```python
fold_column_name = "fold_id"
train_h2o[fold_column_name] = train_h2o["id"] % 10
```

```python
cat_vars = train_h2o.as_data_frame().columns[train_h2o.isfactor()].drop("target").tolist()
```

```python
num_vars = train_h2o.as_data_frame().columns[list(map(lambda x: not x, train_h2o.isfactor()))].drop(["id", "fold_id"]).tolist()
```

```python
X_sklearn_train, X_sklearn_test = np.split(
    pd.get_dummies(pd.concat([train[cat_vars], pd.get_dummies(test[cat_vars])])).fillna(0.),
    [train.shape[0]],
    axis=0)
```

```python
train.shape
```

```python
test.shape
```

```python
X_sklearn_train = pd.concat([X_sklearn_train, train[num_vars]], axis=1)
```

```python
X_sklearn_test = pd.concat([X_sklearn_test, test[num_vars]], axis=1)
```

```python
print(X_sklearn_train.shape, X_sklearn_test.shape)
```

```python
minmax_scaler = MinMaxScaler()
X_sklearn_train_scaled = minmax_scaler.fit_transform(X_sklearn_train)
```

```python
X_sklearn_test_scaled = minmax_scaler.transform(X_sklearn_test)
```

## Baseline

As our baseline we consider a non-optimized logistic regression model. This is the simplest model that we can build (we could also consider a majority class classifier, this approach would achieve high accuracy but very poor auc).

```python
from sklearn.linear_model import LogisticRegression
```

```python
metrics.SCORERS.keys()
```

```python
y_true_train = train_h2o.as_data_frame()["target"].values
lr = LogisticRegression()
print("Baseline CV AUC:", np.mean(cross_val_score(lr, X_sklearn_train, y_true_train, scoring="roc_auc")))
```

## Models

For every model we build there are two validation steps. First we run random grid search for model optimization to find the best set of hyperparameters and after this step is complete we run another cross validation to store predictions (for stacking) and to estimate the final score.

After each models is built, whenever possible we create a variable importance plot to highlight the most important variable for the models. This approach is not always feasible, non-parametric models such as K-NN will not have a feature importance measure and for other models like deep learning models such measure is not straighforward, but whenever it is possible, it is provided.

For each model we also provide the cross validation AUC, which is the expected model performance in the test set. For the ensemble we expect to  AUC to be slightly superior to the AUC of the individual predictions.

```python
X_stack_test = np.zeros((test.shape[0], 6))
```

#### Penalized linear model

```python
glm_fit = H2OGeneralizedLinearEstimator(family='binomial',
                                        model_id='glm_fit',
                                        lambda_search=True,
                                        seed=1,
                                        keep_cross_validation_predictions=True)
```

```python
glm_fit.train(x=x, y=y, training_frame=train_h2o, fold_column=fold_column_name)
```

```python
glm_fit.cross_validation_predictions()[2].shape
```

```python
print("GLM AUC", glm_fit.cross_validation_metrics_summary().as_data_frame().loc[1, "mean"])
```

```python
glm_fit.varimp_plot()
```

```python
df_coef= pd.Series(glm_fit.coef())
df_coef= df_coef[np.abs(df_coef) > 0]
df_coef.head()
```

```python
glm_fit.std_coef_plot(num_of_features=10)
```

```python
ROC_Curve(glm_fit, train_h2o)
```

```python
hyper_parameters = {'alpha': [i * 0.01 for i in range(1, 10)]}
# test search_criteria plumbing and max_models
search_criteria = { 'strategy': "RandomDiscrete",
                   'stopping_metric': "AUC",
                   'max_models': 40,
                   'stopping_tolerance': 0.001,
                    'stopping_rounds': 2,
                  "seed": 1}
glm3 = H2OGridSearch(H2OGeneralizedLinearEstimator(family='binomial',lambda_search = True),
                             hyper_parameters,grid_id="glm3", search_criteria=search_criteria)
glm3.train(x=x, y=y, training_frame=train_h2o, fold_column=fold_column_name)
```

```python
print("GLM AUC", glm3[0].cross_validation_metrics_summary().as_data_frame().loc[1, "mean"])
```

```python
X_stack_test[:, 0] = glm_fit.predict(test_h2o).as_data_frame()["p0"]
```

#### Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score
```

```python
dt = DecisionTreeClassifier()
```

```python
dt_param_grid = {"max_depth" : [3, 5, 7, 9, 11, 13],
                 "min_samples_leaf" : [1, 5, 10, 20, 50, 100]}
```

```python
dt_grid_search_cv = RandomizedSearchCV(dt, dt_param_grid, n_iter=10, scoring="roc_auc")
```

```python
dt_grid_search_cv.fit(X_sklearn_train, y_true_train)
```

```python
print("Decision Tree AUC: ", dt_grid_search_cv.best_score_)
```

```python
dt_grid_search_cv.best_params_
```

```python
best_vars = np.argsort(dt_grid_search_cv.best_estimator_.feature_importances_)[::-1][:10]
var_imp = sorted(dt_grid_search_cv.best_estimator_.feature_importances_)[::-1][:10]
best_var_names = X_sklearn_train.columns[best_vars]

fig, ax = plt.subplots(dpi=100)
sns.barplot(best_var_names, var_imp, ax=ax)
ax.set_title("Decision tree variable importance")
fig.autofmt_xdate()
```

```python
X_stack_test[:, 1] = dt_grid_search_cv.best_estimator_.predict_proba(X_sklearn_test)[:, 0]
```

#### Nearest neighbors

K-NN is a non-parametric algorithm that predicts the class of a example based on the class of the most similar samples in the dataset. It is common to use a distance metric such as euclidian distance to measure similarity between samples and in classification problems a voting approach is used to decide which class the example will belong to. The K in K-NN is the number of similar samples used in the voting procedure and is a hyperparameter of the algorithm that needs to be validated.

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier()
```

```python
knn_param_grid = {"n_neighbors" : [3, 5, 7, 9, 11, 13]}
```

```python
knn_grid_search_cv = RandomizedSearchCV(knn, knn_param_grid, n_iter=5, scoring="roc_auc")
```

```python
knn_grid_search_cv.fit(X_sklearn_train_scaled, y_true_train)
```

```python
print("K-NN AUC", knn_grid_search_cv.best_score_)
```

```python
knn_grid_search_cv.best_params_
```

```python
X_stack_test[:, 2] = knn_grid_search_cv.best_estimator_.predict_proba(X_sklearn_test_scaled)[:, 0]
```

#### Gradient Boosting

Gradient boosting is a ensemble algorithm that builds a series of models (week learners) sequentially with each new model trying to correct the error of the sum of the already built models. 
As a week learner it is common to use a decision tree as a estimator.

```python
criteria = {"strategy": "RandomDiscrete", "max_models" : 10,
            "max_runtime_secs": 28800, "seed": 1234}
```

```python
gbm_params = {'learn_rate': [0.01, 0.05, 0.1, 0.5], 
                'max_depth': [3, 5, 9, 11],
                'sample_rate': [0.8, 0.9, 1.0],
                'col_sample_rate': [0.2, 0.5, 0.8, 1.0]}
```

```python

gbm_grid = H2OGridSearch(model=H2OGradientBoostingEstimator,
                         grid_id='gbm_grid',
                         hyper_params=gbm_params,
                         search_criteria=criteria)
gbm_grid.train(x=x, y=y, 
                training_frame=train_h2o, 
                fold_column=fold_column_name, 
                ntrees=100,
                seed=1)
```

```python
print("GBM AUC:", gbm_grid.models[0].cross_validation_metrics_summary().as_data_frame().loc[1, "mean"])
```

```python
gbm_grid.models[0].varimp_plot()
```

```python
X_stack_test[:, 3] = gbm_grid.models[0].predict(test_h2o).as_data_frame()["p0"]
```

#### RandomForest

```python
rf_params = {'max_depth': [3, 5, 9, 11],
                'sample_rate': [0.8, 0.9, 1.0],
                'col_sample_rate_per_tree': [0.2, 0.5, 0.8, 1.0]}
```

```python
rf_grid = H2OGridSearch(model=H2ORandomForestEstimator,
                         grid_id='rf_grid',
                         hyper_params=rf_params,
                         search_criteria=criteria)
rf_grid.train(x=x, y=y, 
                training_frame=train_h2o, 
                fold_column=fold_column_name, 
                ntrees=250,
                seed=1)
```

```python
print("Random Forest AUC", rf_grid.models[0].cross_validation_metrics_summary().as_data_frame().loc[1, "mean"])
```

```python
rf_grid.models[0].varimp_plot()
```

```python
X_stack_test[:, 4] = rf_grid.models[0].predict(test_h2o).as_data_frame()["p0"]
```
