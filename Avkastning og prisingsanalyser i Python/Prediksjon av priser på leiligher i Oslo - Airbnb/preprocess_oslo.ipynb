# Cleaning dataset for Modelling 

```python
oslo = pd.read_csv("data/oslo_listings2.csv")
```

```python
oslo.info()
```

```python
list(oslo.columns)
```

```python
# Selecting futures we want to take with us
selected_features = [u'id',
                     u'name',
                     u'price',
                     u'number_of_reviews',
                     u'square_feet',
                     u'first_review',
                     u'accommodates',
                     u'host_response_time',
                     u'bathrooms',
                     u'bedrooms',
                     u'beds',
                     u'security_deposit', 
                     u'cleaning_fee', 
                     u'guests_included',
                     u'extra_people',
                     u'minimum_nights', 
                     u'maximum_nights',
                     u'guests_included',
                     u'availability_365',
                     u'neighbourhood',
                     u'longitude',
                     u'latitude',
                     u'review_scores_rating',
                     u'review_scores_cleanliness',
                     u'review_scores_checkin',
                     u'review_scores_communication', 
                     u'review_scores_location',
                     u'review_scores_value', 
                     u'house_rules',
                     u'amenities',
                     'bed_type', 
                     'room_type',
                     'cancellation_policy',
                     'property_type']
oslo = oslo.loc[:, selected_features]
oslo = oslo.apply(lambda x:x.fillna(x.value_counts().index[0]))
```

```python
oslo.head(2)
```

```python
oslo["review_scores_rating"].describe()
```

```python
oslo["price"] = oslo["price"].str.replace("\$|,", "").astype(float)
oslo["security_deposit"] = oslo["security_deposit"].str.replace("\$|,", "").astype(float)
oslo["cleaning_fee"] = oslo["cleaning_fee"].str.replace("\$|,","").astype(float)
oslo["extra_people"] = oslo["extra_people"].str.replace("\$|,","").astype(float)
oslo["availability"] = oslo["availability_365"] / 365
```

```python
# house rules
#house_rules = oslo["house_rules"].str.lower()
#smoking = house_rules.str.contains("smoke|smoking", na = False)
#oslo.loc[:, "smoking"] = -smoking
        
#pet =  house_rules.str.contains("pet", na = False)
#oslo.loc[:, "pet"] = - pet

#party = house_rules.str.contains("party|parties", na = False)
#oslo.loc[:, "party"] = - party

#guest = house_rules.str.contains("guest|guests", na = False)
#oslo.loc[:, "guest"] = -guest
```

```python
# drop house rules column
oslo = oslo.drop(["house_rules"], axis = 1)
```

```python
# facilites 
oslo["amenities"][1]
```

```python
import re
amenities = list(oslo["amenities"])
total = ",".join(amenities)
total = total.replace("{", "").replace("}","").replace("\"","").split(",")
amenity_items = list(set(total))
amenity_items = list(filter(None, amenity_items))
for i in amenity_items:
    if re.match(r'translation', i):
        amenity_items.remove(i)
```

```python
# which neighbourhood are there in the dataset
oslo["neighbourhood"].unique()
```

```python
# create matrix for amenities
from sklearn.preprocessing import LabelEncoder, OneHotEncoder

table = pd.DataFrame(index = oslo.reset_index().values[:,0], 
                    columns = amenity_items).fillna(0)

# populate amenities - futures
for i in range(len(amenities)):
    for j in amenity_items:
        if j in amenities[i]:
            table.at[i,j] = 1
```

```python
# create table for neighbourhood
col = list(oslo["neighbourhood"])
table_n = pd.DataFrame(index = oslo.reset_index().values[:,0],
                    columns = set(col)).fillna(0)

# populate the matrix
for i in range(len(col)):
    for j in col:
        if j in col[i]:
            table_n.at[i,j] = 1
```

```python
sum_table = np.array(table.sum())
ind = (-sum_table).argsort()[:60]
common_amenities = list(table.sum().iloc[ind].index)

# add the amenities matrix
oslo = oslo.drop(['amenities'], axis = 1)
oslo = pd.concat([oslo, table[common_amenities]], axis = 1)

# add the neighbourhood matrix
oslo = oslo.drop(["neighbourhood"], axis = 1)
oslo = pd.concat([oslo, table_n], axis = 1)
```

```python
# iterate the same process over bed_type, room_type, cancellation_poloty, property_type and host_response_time
columns = ['bed_type', 'room_type', 'cancellation_policy', 'property_type', 'host_response_time']
for column in columns:
    unique_values = list(oslo[column].unique())
    column_list = list(oslo[column])
    new_table = pd.DataFrame(index = oslo.reset_index().values[:,0], columns = unique_values).fillna(0)
    
    for i in range(len( column_list )):
        for item in unique_values:
            if item in column_list[i]:
                new_table.at[i, item] = 1
    oslo = pd.concat( [oslo, new_table], axis = 1)
    oslo = oslo.drop([column], axis = 1)    
```

```python
oslo.info()
```

```python
# write cleaned file to new file - modelling 
oslo.to_csv("data/oslo_clean.csv")
```
