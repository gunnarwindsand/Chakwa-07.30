# Demand Modeling 

In this part we turn our problem up side down, and aim to estimate a demand for the individual listings on airbnb. during this work, we had to make some **assumptions:**

1. **We do not look at the time variable.** listings at airbnb are adjusted for seasonallity as well, we choose to avide this  problem as we did not hold any relevant data over time, so any modelling would be purely speculative. 

2. demand estimation is based on **number of reviews** and listing time. we therefor assume that a listing with a higher number of monthly reviews is more popular, (maybe seen as psycologically more trustworthy than a low -review counterpart) than a similar listing with fewer reviews. Furthermore we adjust this number by assuming that about 20% of the people staying are actually leavning a review and adjust for market growth. 

3. **NOTE DOWN FURTHER ASSUMPTIONS**



```python
# import packages
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.formula.api import ols

plt.style.use("ggplot")
```

```python
df = pd.read_csv("data/oslo_clean.csv", index_col = [0])
df2 = pd.read_csv("data/oslo_clean_visualization.csv")[["neighbourhood", "id"]]
df = pd.merge(df, df2, on = "id")
df = df[df["price"] < 2000]
df = df[df["price"] > 200]
df.head(2)
```

```python
# select the variables we want to work on
selected_feautures = ["id",
                      "price",
                      "first_review",
                      "number_of_reviews",
                      "neighbourhood"]

df = df.loc[:, selected_feautures]
```

```python
df["first_review"] = pd.to_datetime(df["first_review"])
df["year"] = df["first_review"].dt.year
df["month"] = df["first_review"].dt.month
df["day"] = df["first_review"].dt.day

df["months_listed"] = (2019 - df["year"])*12
df["days_listed"] = df["months_listed"]*30
df.head(2)
```

```python
# Fill 0 reviews with random number between 1 and 4
new_reviews = list()
for i in df["number_of_reviews"]:
    if i  == 0:
        new_reviews.append(np.random.randint(1,5))
    else:
        new_reviews.append(i)
```

```python
# mean occupation rate is 14-16% in similar nordic cities
mean_occ = 0.15 # proxy yearly occupancy rate (sweden and denmark)
adj = 0.5 # adjust for people not placing reviews - assume 20% place a review
growth = .10 #general market growth
avg_days = 3

# populate number of reviews with new revies (no zero values)
df["number_of_reviews"] = new_reviews
df["rev_per_month"] = round(df["number_of_reviews"]/df["months_listed"], 4)
df["rev_per_day"] = round(df["number_of_reviews"]/df["days_listed"], 4)

# estimate motnly demand
df["est_monthly_demand"] = round((df["number_of_reviews"]/adj)*avg_days/df["months_listed"]*(1+growth))

# adjust for unaturally high montly demand
maxVal = 25
df["est_monthly_demand"] = df['est_monthly_demand'].where(df['est_monthly_demand'] <= maxVal, maxVal)
# remove columns with no relevant data
df = df[df["year"] < 2019]
```

## Filter per neighborhood

```python
def sort(x):
    gr = pd.DataFrame(x.groupby(["price"])["est_monthly_demand"].count().reset_index())
    gr["demand"] = np.sum(gr["est_monthly_demand"]) - np.cumsum(gr["est_monthly_demand"])
    gr["root"] = np.sqrt(gr["demand"])
    gr = gr[~gr.isin([np.nan, np.inf, -np.inf]).any(1)]
    return gr
```

```python
alna = sort(df[df["neighbourhood"] == "Alna"])
bjerke = sort(df[df["neighbourhood"] == "Bjerke"])
frogner = sort(df[df["neighbourhood"] == "Frogner"])
gamle_oslo = sort(df[df["neighbourhood"] == "Gamle Oslo"])
grefsen = sort(df[df["neighbourhood"] == "Greften"])
grunerlokka = sort(df[df["neighbourhood"] == "Grünerløkka"])
major = sort(df[df["neighbourhood"] == "Majorstuen"])
nordre_aker = sort(df[df["neighbourhood"] == "Nordre Aker"])
nordstrand = sort(df[df["neighbourhood"] == "Nordstrand"])
sagene = sort(df[df["neighbourhood"] == "Sagene"])
sentrum = sort(df[df["neighbourhood"] == "Sentrum"])
st_haugen = sort(df[df["neighbourhood"] == "St.Hanshaugen"])
ullern = sort(df[df["neighbourhood"] == "Ullern"])
ostensjo = sort(df[df["neighbourhood"] == "Østensjø"])
```

```python
def faen(x):
    plt.scatter(x = x["root"], y = x["price"], label = "x")
    
plt.figure(figsize=(12,6))
faen(alna)
faen(bjerke)
faen(gamle_oslo)
faen(grefsen)
faen(grunerlokka)
faen(major)
faen(nordre_aker)
faen(nordstrand)
faen(sagene)
faen(sentrum)
faen(st_haugen)
faen(ullern)
faen(ostensjo)


plt.xlabel("$\sqrt{Demand}$")
plt.ylabel("Price")
plt.title("Demand Slopes for the respective Neighbourhoods in Oslo")
plt.show()
```

## Decision Model 

```python
# merge the predicted dataframe with our estimated monthly demand
predicted_price = pd.read_csv("data/predicted_results.csv", index_col = [0])
#predicted_price.rename(columns ={"ID":"id"}, inplace = True)
df["ID"] = df.index
df_new = pd.merge(predicted_price, df, on ="ID" )
df_new.head()
```

let's take ID **4381541**, which have an recommended price of 534 NOK, but actually charges 346 NOK per day. our estimated monthly demand at $P = 346$ is 17 days a month. 

```python
def change_demand(old_price, new_price, est_monthly_demand, data):
    
    """
    Summary:
    Estimation of change in demand by changing price given in 
    previous modelling. assumes market growth, 365 days 
    availability 
    -------------------------------------------------------------
    params:
    old_price: price before modeling (modelling.ipnb)
    new_price: price returned from pricing-model
    est_monthly_demand: estimated demand based on popularity
    data: data query - e.g., frogner, major, etc
    -------------------------------------------------------------
    Returns:
    estimated optimized revenue and old revenue  
    """
    # fit a OLS model to the data
    model = ols("root ~ price", data = data)
    result = model.fit()
    
    # calculate new and old demand based on linear regression
    demand_new = (result.params[0] + (result.params[1]*new_price))**2
    demand_old = (result.params[0] + (result.params[1]*old_price))**2
    
    # return global demand change
    print("New gloabl demand",round(demand_new, 2))
    print("Old global demand",round(demand_old, 2))
    
    # calcualte change in profit
    profit_new = demand_new*new_price
    profit_old = demand_old*old_price
    #print("New profit", profit_new)
    #print("Old Profit", profit_old)
    
    # price elastisity
    c_p = (new_price - old_price)/((old_price + new_price)/2)*100
    c_q = (demand_new - demand_old)/((demand_old + demand_new)/2)*100 
    ela = abs(c_q / c_p)
    
    # percentage change in price
    p_c = abs((new_price - old_price) / old_price)
    
    if new_price > old_price:
        ch = 1 - (p_c * ela)
    else:
        ch = 1 + (p_c * ela)
    
    # adjust montly demand
    adj_d = (est_monthly_demand * ch) 
    
    # new and old profit for listing X
    pi = (adj_d * new_price)
    old = (est_monthly_demand*old_price)
    
    if pi > old:
        print("Change Price!")
        print("Expexted Revenue", round(pi, 2), "NOK")
        print("New estimated demand",round(adj_d, 0), "days")
        print("Old Revenue", round(old, 2), "NOK")
    else:
        print("Keep Old Price")
        print("Revenue", round(old,2), "NOK")
```

```python
def change_demand(old_price, new_price, est_monthly_demand, data):
    """
    Summary:
    Estimation of change in demand by changing price given in 
    previous modelling. assumes market growth, 365 days 
    availability 
    -------------------------------------------------------------
    params:
    old_price: price before modeling (modelling.ipnb)
    new_price: price returned from pricing-model
    est_monthly_demand: estimated demand based on popularity
    data: data query - e.g., frogner, major, etc
    -------------------------------------------------------------
    Returns:
    estimated optimized revenue and old revenue
    """
    old_price = int(old_price)
    if new_price > old_price:
        new_price = int(new_price + ((new_price/100)*20))
    est_monthly_demand = est_monthly_demand
    
    # initialize lists 
    output = list()
    dem = list()
    price_list = list()
    
    if old_price < new_price:
        
        # iterate over all the price-points
        for i in range(old_price, new_price, 1):

            # fit OLS model on data query
            model = ols("root ~ price", data = gamle_oslo)
            result = model.fit()

            #calcutate the global demands
            demand_new = (result.params[0] + (result.params[1]*i))**2
            demand_old = (result.params[0] + (result.params[1]*old_price))**2
            #print("New gloabl demand",round(demand_new, 2))
            #print("Old global Demand", round(demand_old,2))

            # calculate gloabal profits
            profit_new = demand_new * i
            profit_old = demand_old*old_price

            # demand appriximation
            ch = (demand_new - demand_old)/demand_old
            if ch < 0:
                ch = 1 - abs(ch)
            else:
                ch = 1 + ch 

            # adjust montly demand and assign to list
            adj_d = (est_monthly_demand * ch)
            dem.append(adj_d)

            # Adjusted revenue and assign to list
            pi = (adj_d * i)
            output.append(pi)
            
            # append prices to list
            price_list.append(i)

            # price elastisity
            c_p = (new_price - old_price)/((old_price + new_price)/2)*100
            c_q = (demand_new - demand_old)/((demand_old + demand_new)/2)*100 
            ela = abs(c_q / c_p)    

        # display results
        if ela > 1:
            print("Elastic", ela)
        elif ela == 0:
            print("Unitary", ela)
        else:
            print("Unelastic", ela)
    
    else:
       
        # iterate over all the price-points
        for i in range(new_price, old_price, 1):

            # fit OLS model on data query
            model = ols("root ~ price", data = gamle_oslo)
            result = model.fit()

            #calcutate the global demands
            demand_new = (result.params[0] + (result.params[1]*i))**2
            demand_old = (result.params[0] + (result.params[1]*old_price))**2
            #print("New gloabl demand",round(demand_new, 2))
            #print("Old global Demand", round(demand_old,2))

            # calculate gloabal profits
            profit_new = demand_new * i
            profit_old = demand_old*old_price

            # demand appriximation
            ch = (demand_new - demand_old)/demand_old
            if ch < 0:
                ch = 1 - abs(ch)
            else:
                ch = 1 + ch 

            # adjust montly demand and assign to list
            adj_d = (est_monthly_demand * ch)
            dem.append(adj_d)

            # Adjusted revenue and assign to list
            pi = (adj_d * i)
            output.append(pi)
            
            # append prices to list
            price_list.append(i)

            # price elastisity
            c_p = (new_price - old_price)/((old_price + new_price)/2)*100
            c_q = (demand_new - demand_old)/((demand_old + demand_new)/2)*100 
            ela = abs(c_q / c_p)    

        # display results
        if ela > 1:
            print("Elastic", ela)
        elif ela == 0:
            print("Unitary", ela)
        else:
            print("Unelastic", ela)

    # old revenue 
    old = (est_monthly_demand*old_price)
    new = (adj_d * new_price)
    # assign results to dataframe
    d = pd.DataFrame({"demand" : dem, "revenue": output, "Optimal Price":price_list})
    
    # initialize if E(revenue|$\Delta$price) > old revenue
    
    if  :
        print("Change to new price")
        print(new)
    
    else:
        print("Keep Old price")
        print(old)
    
    print("optimal price is")
    print(d[d["revenue"] == d["revenue"].max()])
    
        
    if old_price < new_price:
        # visualize
        plt.figure(figsize = (12,5)) 
        plt.plot(range(old_price,new_price,1), output)
        plt.xlabel("Price")
        plt.ylabel("Revenue")
        # old revenue 
        plt.text(y = old+5, s = "Old revenue", x = new_price- 25)
        plt.axhline(y = old, c = "black", alpha = 0.7, ls = "--")
        plt.legend(["Projected Revenue"])
        plt.show()
    else:
        plt.figure(figsize = (12,5)) 
        plt.plot(range(new_price,old_price,1), output)
        plt.xlabel("Price")
        plt.ylabel("Revenue")
        # old revenue 
        plt.text(y = old+5, s = "Old revenue", x = new_price- 25)
        plt.axhline(y = old, c = "black", alpha = 0.7, ls = "--")
        plt.legend(["Projected Revenue"])
        plt.show()        
```
